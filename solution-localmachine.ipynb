```python
p_class = pd.get_dummies(x_train['Pclass'],prefix='Pclass')
```

```python
x_train = x_train.drop('Pclass',axis=1)
```

```python
x_train = pd.concat([x_train,p_class],axis=1)
# x_train = x_train.join(p_class)
```

```python
x_train.describe()
```

```python
sex = pd.get_dummies(x_train['Sex'],prefix='Sex')
```

```python
x_train=x_train.drop('Sex',axis=1)
```

```python
x_train = pd.concat([x_train,sex],axis=1)
```

```python
# x_train = x_train.drop_duplicates()
x_train.describe()
```

```python
mean_age = 29.699118
x_train['Age'].fillna(value=mean_age, inplace = True)
x_train['Age'].isnull().sum()
```

```python
x_train['Age'] /= mean_age
x_train['Age']
```

```python
print(x_train.describe())
print(x_train.columns)
```

```python
x_train.isnull().sum()
```

```python
x_train = x_train.drop(['Cabin'],axis=1)
x_train.columns
```

```python
x_train['Embarked'].value_counts()
```

```python
x_train['Embarked'].fillna(value='S', inplace = True)
```

```python
embarked = pd.get_dummies(x_train['Embarked'],prefix='Emb')
embarked.head()
```

```python
x_train=x_train.drop(['Embarked'],axis=1)
x_train = pd.concat([x_train,embarked],axis=1)
x_train.head()
```

```python
x_train.describe()
```

```python
x_train=x_train.drop(['Ticket','PassengerId'],axis=1)
```

```python
x_train['Fare'].fillna(value=51.758668, inplace = True)
x_train.isnull().sum()
```

```python
mean_fare = 32.204208
x_train['Fare'] /= mean_fare
x_train.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_train,y_train, test_size = 0.2, random_state = 0)
```

```python
# x = x_train
# X_train = x.iloc[:891]
# X_test = x.iloc[891:]
```

```python
from sklearn import svm, datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
```

```python
model1 = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial')
model1.fit(X_train,y_train);
predicted1 = model1.predict(X_test);
print("Logistic Reg",accuracy_score(y_test,predicted1)*100,"%")

model2 = svm.SVC(kernel='linear' ,C=1,gamma=1);
model2.fit(X_train,y_train);
model2.score(X_train,y_train);
predicted2 = model2.predict(X_test);
print("SVM",accuracy_score(y_test,predicted2)*100,"%")

model3 = KNeighborsClassifier(n_neighbors=2)
model3.fit(X_train,y_train)
predicted3 = model3.predict(X_test)
print("KNN",accuracy_score(y_test,predicted3)*100,"%")

model4 = RandomForestClassifier(n_estimators=100,random_state=50)
model4.fit(X_train,y_train)
predicted4 = model4.predict(X_test)
print("RandomForest",accuracy_score(y_test,predicted4)*100,"%")

model5 = DecisionTreeClassifier()
model5.fit(X_train,y_train)
predicted5 = model5.predict(X_test)
print("Decision Tree",accuracy_score(y_test,predicted5)*100,"%")

model6 = MLPClassifier(solver='lbfgs', alpha=1e-20, hidden_layer_sizes=(5, 3), random_state=1)
model6.fit(X_train,y_train)
predicted6 = model6.predict(X_test)
print("Neural Net",accuracy_score(y_test,predicted6)*100,"%")

```
